```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as  plt 

pages = pages = ["https://www.kammi.it/scarpe-donna/sneakers/",
                 "https://www.kammi.it/scarpe-donna/sneakers/pag2/"]

dict = {"modello": [], "descrizione": [], "prezzo": []}

for url in pages:
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        modelli = soup.find_all("p", attrs={"class": "product-title"})
        descrizioni = soup.find_all("p", attrs={"class": "product-description"})
        prezzi = soup.find_all("p", attrs={"class": "product-price"})
        
        prezzi_clean = [prezzo.text for i, prezzo in enumerate(prezzi) if i % 2 != 0]
        
        for modello, descrizione, prezzo in zip(modelli, descrizioni, prezzi_clean):
            dict["categoria"] = "sneakers"
            dict["modello"].append(modello.text.strip())
            dict["descrizione"].append(descrizione.text.strip())
            dict["prezzo"].append(prezzo.replace("€","").replace(",",".").strip())

df1= pd.DataFrame(dict)
df1
```
